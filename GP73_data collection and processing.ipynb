# Data Visualisation

```python



# Calculate descriptive statistics for the satisfaction level
stats = df['satisfaction_level'].describe()

# Extract relevant statistics
mean = stats['mean']
std = stats['std']
q1 = stats['25%']
q2 = stats['50%']
q3 = stats['75%']
min_val = stats['min']
max_val = stats['max']

# Create a bar chart for the statistics
plt.bar(['Mean', 'Std', 'Min', 'Q1', 'Q2', 'Q3', 'Max'], [mean, std, min_val, q1, q2, q3, max_val])

# Set the title and labels
plt.title('Satisfaction Level Descriptive Statistics Bar Chart')
plt.xlabel('Statistics')
plt.ylabel('Value')

# Show the chart
plt.show()
```

```python

sns.histplot(df['satisfaction_level'])
```

```python
sns.scatterplot(data=df, x='satisfaction_level', y='last_evaluation')
```

```python
# check for outliers
sns.boxplot(data=df, x='satisfaction_level')
```
